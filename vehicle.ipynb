```python
import pandas as pd

df=pd.read_csv('vehicle_silhouette_dataset.csv')
df.head()
```

```python
updated_df=df.drop(columns='vehicle_class')
updated_df.head()
```

```python
columns_list=[]
for i in updated_df.columns:
    columns_list.append(i)
columns_list
```

```python
updated_df.info()
```

```python
updated_df.describe()
```

```python
total = updated_df.isnull().sum().sort_values(ascending=False)
total
```

```python
updated_df.astype('str').dtypes
```

```python
updated_df
```

```python
from sklearn.preprocessing import MinMaxScaler
normalize_df=pd.DataFrame([])
values=updated_df.values
min_max_scaler=MinMaxScaler()
scaled=min_max_scaler.fit_transform(values)
normalize_df=pd.DataFrame(scaled)
normalize_df.columns=columns_list
normalize_df

```

```python
target_set=set(df['vehicle_class'])
target_set
```

```python
target_list=[]
targetcol_list=list(df['vehicle_class'])
for i in targetcol_list:
    if i=='bus':
        target_list.append(0)
    elif i=='opel':
        target_list.append(1)
    elif i=='saab':
        target_list.append(2)
    elif i=='van':
        target_list.append(3)
target_list
    
    
```

```python
target_df=pd.DataFrame({'vehicle_name':target_list})
target_df
```

# Training = 70 Testing = 30

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(normalize_df,target_df,test_size=0.3,random_state=42)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
ypred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,ypred)
acc
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.naive_bayes import GaussianNB
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)
y_predict= random_forest.predict(x_test)
randomforest_accuracy=accuracy_score(y_test, y_predict)
print("Random Forest Accuarcy = ",randomforest_accuracy)
```

```python
logreg=LogisticRegression()
logreg.fit(x_train, y_train)
y_predict=logreg.predict(x_test)
logreg_accuracy=accuracy_score(y_test, y_predict)
print("Logistic Regression Accuarcy = ",logreg_accuracy)
```

```python
perceptron=Perceptron(max_iter=5)
perceptron.fit(x_train, y_train)
y_predict=perceptron.predict(x_test)
perceptron_accuracy=accuracy_score(y_test, y_predict)
print("Perceptron Accuarcy = ",perceptron_accuracy)
```

```python
gaussian=GaussianNB() 
gaussian.fit(x_train, y_train)  
y_predict=gaussian.predict(x_test)
guassian_accuracy=accuracy_score(y_test, y_predict)
print("Guassian Accuarcy = ",guassian_accuracy)
```
